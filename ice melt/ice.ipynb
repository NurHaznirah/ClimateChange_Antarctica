```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_2009= pd.read_excel('C:/Users/User/Documents/INTERNSHIP/eda data/ice melt/2009.xlsx')
```

```python
df_2019= pd.read_excel('C:/Users/User/Documents/INTERNSHIP/eda data/ice melt/2019.xlsx')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_2009['THICK'], bins=30, kde=True)
plt.title('Distribution of Ice Thickness')
plt.xlabel('Ice Thickness')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='LON', y='LAT', data=df_2009, hue='THICK', palette='viridis', size='THICK', sizes=(10, 200))
plt.title('Ice Thickness Distribution by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.lineplot(x='TIME', y='THICK', data=df_2009)
plt.title('Ice Thickness Over Time')
plt.xlabel('Time')
plt.ylabel('Ice Thickness')
plt.show()
```

```python
plt.plot(df_2009['THICK'], df_2009['TIME'], label='2009', marker='o', linestyle='-')
```

```python
plt.plot(df_2019['TIME'], df_2019['THICK'], label='2019', marker='o', linestyle='-')
```

```python
plt.xlabel('Time')
plt.ylabel('Ice Thickness')
plt.title('Ice Thickness Over Time (2009 and 2019)')

# Explicitly add legend with labels
plt.legend(labels=['2009', '2019'])

plt.show()
```
