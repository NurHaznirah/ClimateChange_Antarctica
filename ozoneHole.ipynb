```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
data = pd.read_excel('D:/competition/Ozone_hole.xlsx')
data.head()
```

```python
data.shape
```

This dataset has 42 rows and 3 columns.

```python
data.info()
```

```python
null_values = data.isnull().sum()
print(null_values)
```

```python
sns.lineplot(x='Year', y='Hole Area', data=data)
plt.show()
```

Based on the graph above, we can see that as the years passed, the hole area which is ozone layer in Antarctica is increase. This is because of the increased UV radiation reached the earth surface in Antarctica thus, make the ozone layer become thin.

```python
sns.lineplot(x='Year', y='Minimum Ozone', data=data)
plt.show()
```

Based on the graph above, the minimum ozone refers to the minimum level of ozone concentration where the minimum ozone is decline as the year passed 

```python
sns.scatterplot( x="Year", y='Minimum Ozone', data=data, hue='Hole Area')
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```

```python
plt.hist(data['Minimum Ozone'], bins=5)
plt.show()
```
