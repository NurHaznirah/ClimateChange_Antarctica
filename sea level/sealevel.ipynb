```python
df['date'] = pd.to_datetime(df['date'])

# Assuming you have 'dates' and 'mmfrom1993-2008average' as your data
plt.figure(figsize=(15, 6))
plt.plot(df['date'], df['mmfrom1993-2008average'], marker='o', linestyle='-', label='mmfrom1993-2008average')

plt.xlabel('Date', fontsize=12)
plt.ylabel('mm from 1993-2008 average', fontsize=12)
plt.title('Sea Levels Over Time', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

Between 1880 and 2020, global sea levels have been observed to rise.
